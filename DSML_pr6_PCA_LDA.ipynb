```python
import pandas as pd
dataset = pd.read_csv('/content/Wine.csv')
dataset.corr()
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=0)
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

#PCA
#(UN-Supervised Feature Extration dimensionality reduction technique)

PCA - Principal Component Analysis



Convert 13 column into 2 columns.. for that we use n_components = number

Provide matrix for that define or considered 2 columns.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
X_train
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
cm
```

# LDA
# (Supervised Feature Extration dimensionality reduction technique)

LDA = Linear Discriminant Analysis

```python
import pandas as pd
dataset = pd.read_csv('Wine.csv')
dataset.corr()
```

```python
X = dataset.iloc[:,:-1].values
Y = dataset.iloc[:,-1].values


from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2, random_state=0)


from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
X_train = lda.fit_transform(X_train,Y_train)
X_test = lda.transform(X_test) 
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0)
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,y_pred)
cm
```
